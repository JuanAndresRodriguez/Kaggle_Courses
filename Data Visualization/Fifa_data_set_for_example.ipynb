```python
import pandas as pd
data = pd.read_csv("fifa_ranking.csv")
```

```python
data.head()
```

```python
countries = ["ARG", "BRA", "ESP", "FRA", "GER", "ITA"]
data = data[data["country_abrv"].isin( countries)]
# data.query(country_abrv in ["ARG", "BRA", "ESP", "FRA", "GER", "ITA"])
```

```python
data = data[["rank_date", "country_abrv", "rank"]]
data.head()
pivoted = data.pivot_table(index='rank_date', columns='country_abrv', values='rank')#.reset_index()
pivoted.columns.name=None
print(pivoted)
```
